This notebook is to get a feeling of how the KL divergence behaves

Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

Functions to calculate KL divergence and plot hisograms 

```python
def plot_dists(dist_1, dist_2):
    _, _, _ = plt.hist(dist_1,60, density=True, alpha=0.5)
    _, _, _ = plt.hist(dist_2,60, density=True, alpha=0.5)
# Calculate the KL Divergence
def KL_divergence(p, q):
    D = 0
    for i in range(len(p)):
        if p[i] > 0 and q[i] > 0:
            D += p[i] * np.log(p[i]/q[i])
        elif p[i] > 0 and q[i] == 0:
            D += q[i]

    return D
```

Sample two normal distributions and plot their histograms and calculate their KL divergence

```python
# numpy normal distribution
mu, sigma = 5, 0.1 # mean and standard deviation
dist_normal1 = np.random.normal(mu, sigma, 1000)
dist_normal2 = np.random.normal(mu+0.1, sigma, 1000)

plot_dists(dist_normal1, dist_normal2)
print("KL-divergence: ", KL_divergence(dist_normal1, dist_normal2))
```

If they are the same, the KL divergence should be 0

```python
plot_dists(dist_normal1, dist_normal1)
print("KL-divergence: ", KL_divergence(dist_normal1, dist_normal1))
```

If they are further apart, the KL divergence should be higher

```python
# numpy normal distribution
mu, sigma = 5, 0.1 # mean and standard deviation
dist_normal1 = np.random.normal(mu, sigma, 1000)
dist_normal2 = np.random.normal(mu+0.3, sigma, 1000)

plot_dists(dist_normal1, dist_normal2)
print("KL-divergence: ", KL_divergence(dist_normal1, dist_normal2))
```
